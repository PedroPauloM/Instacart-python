```python
#Deriving Columns
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged['loyalty_flag'].value_counts(dropna= False)
```

```python
# Develop a spending flag for each user based on the average price of products in the desired orders
```

```python
ords_prods_merged['mean_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merged.head(15)
```

```python
# Create flags to assign "spending" labels to user ID based on mean product price
```

```python
ords_prods_merged.loc[ords_prods_merged['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequencies of spending flags 
```

```python
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
# Implement an order frequency indicator that reveals users' buying behavior through their orders.
```

```python
ords_prods_merged['median_days'] = ords_prods_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Create flags to assign "freqency" labels to user ID based on median days since last order
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_days'] > 10) & (ords_prods_merged['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking frequencies of spending flags
```
